**1.  What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples?**

  -  Anomaly Detection (also called Outlier Detection) is the process of identifying data points, events, or observations that significantly differ from the normal behavior of a dataset. These unusual observations may indicate errors, fraud, faults, intrusions, or rare events.

     Types of Anomalies :

  1. Point Anomalies :

     A point anomaly occurs when a single data instance is significantly different from the rest of the data

     Example :
    
     1. A credit card transaction of ₹5,00,000 when the customer usually spends less than ₹5,000.

     2. A sudden spike in website traffic due to a cyberattack

  2. Contextual Anomalies :

     A contextual anomaly is abnormal only within a specific context. The same observation may be normal in one situation and anomalous in another.

     Example :

     1. A login at 3 AM may be normal for a night-shift employee but anomalous for a regular office worker.

     2. High electricity consumption during the day may be normal, but the same usage late at night may indicate a problem.

  3. Collective Anomalies :

     A collective anomaly occurs when a group of related observations is anomalous, even though individual observations may appear normal

     Example :

     1. Multiple failed login attempts followed by a successful login (possible hacking attempt).

     2. Unusual sequence of sensor readings indicating machine failure.

     3. A sudden burst of network requests during a Distributed Denial-of-Service (DDoS) attack.

**2.   Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases ?**

  - 1. Isolation Forest

     approach  :

     Builds multiple random binary trees.

     Randomly selects features and split values.

     Anomalies are isolated in fewer splits because they are rare and different.


     Suitable Use Cases :

     Credit card fraud detection.

     Network intrusion detection.

     Large-scale industrial monitoring.

     Log analysis.

     2. DBSCAN :

     approach :

     Forms clusters based on point density.

      Uses two parameters:

      eps: neighborhood radius.

      min_samples: minimum points required to form a dense region.

     Points not belonging to any cluster are labeled as noise (anomalies).


     Suitable Use Cases :

     Geographic data analysis.

     GPS trajectory analysis.

     Image segmentation.

     Spatial anomaly detection.

  3. Local Outlier Factor (LOF) :

     approach:

      Compares the local density of a point with the density of its neighbors.

     If a point has much lower density than nearby points, it is considered an outlier.

     Produces an LOF score:
     ≈ 1 → Normal

     1 → Potential outlier

     Much greater than 1 → Strong outlier


     Suitable Use Cases :

     Fraud detection in customer groups.

     Medical anomaly detection.

     Sensor monitoring.

     Intrusion detection where anomalies are local rather than global.

**3.   What are the key components of a Time Series? Explain each with one
example?**

  -  A Time Series is a sequence of observations recorded at regular intervals over time. Time series data is generally composed of four key components:

  1. Trend (T) :

     The trend represents the long-term movement or direction of the data over time. It may be increasing, decreasing, or stable.

     Example:

     The annual sales of an e-commerce company increase from ₹10 lakh in 2020 to ₹25 lakh in 2025

     2. Seasonal Variation (S) :

     Seasonality refers to regular and repeating patterns that occur within a fixed period, such as days, months, quarters, or seasons.

     Example:

     An ice cream shop experiences higher sales during summer and lower sales during winter every year.

     Summer sales: 1000 units/month

     Winter sales: 300 units/month


     3. Cyclical Variation (C) :

     Cyclical variations are long-term fluctuations that occur over periods longer than one year, usually due to economic or business cycles.

     Example:

     The automobile industry may experience:

     High sales during economic growth.

     Reduced sales during recession.

     Recovery afterward.


     4. Irregular or Random Variation (I) :

     These are unpredictable fluctuations caused by unexpected events and cannot be forecast easily.

     Example:

     A sudden pandemic, natural disaster, or strike causing a sharp drop in airline passenger traffic.

**4.  Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?**

  -  A stationary time series is one whose statistical properties—such as its mean, variance, and autocorrelation—remain constant over time. It lacks long-term trends or seasonal patterns, making its behavior predictable and much easier to model.

  1. How to Test for Stationarity :

      Visual Inspection: Plot the data over time. A stationary series will move horizontally around a constant mean with uniform variance, while non-stationary data will show clear trends (upward/downward) or changing volatility.

      Augmented Dickey-Fuller (ADF) Test: The most common statistical test. The null hypothesis assumes the data is non-stationary (contains a unit root). If the p-value is less than your significance level, you can reject the null hypothesis and conclude the series is stationary

      KPSS Test: Tests for trend-stationarity. The null hypothesis assumes the data is stationary. It is often used alongside the ADF test to confirm results.

  2. How to Transform a Non-Stationary Series

     Differencing: Subtracting the previous observation from the current observation. This effectively removes trends and seasonality, stabilizing the mean.

     Log Transformation: Applying a natural log to the data can stabilize variance, especially in series that exhibit exponential growth or widening fluctuations over time.

     Detrending & Seasonal Adjustment: Using mathematical techniques like rolling mean subtraction or classical seasonal decomposition to extract and remove underlying trends and repeating seasonal cycles.

**5.   Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application?**

  -  These time series models are progressive extensions of one another, adding layers of complexity to capture real-world data patterns. They scale from simple historical predictions (AR/MA) to complex, multi-variable systems with trends and seasonality (SARIMAX)

  1. AR  ModelStructure: Predicts future values based solely on its own past values (lags). The forecast is a linear combination of past observations plus a random error term

     Application: Used when past values have a direct, linear correlation with future values. Example: Financial forecasting where a stock's price today depends heavily on its price yesterday.

  2. MA  ModelStructure: Predicts future values based on past forecast errors (the difference between the actual value and the predicted value) rather than past data points

     Application: Used to smooth out data and filter out short-term fluctuations/shocks. Example: Analyzing retail sales where a sudden, unexpected drop in a previous month's sales corrects itself in the following months.

  3. ARIMA Structure: Combines AR, MA, and an 'Integrated' (I) component. The 'I' component applies differencing (subtracting the previous time step's value) to make non-stationary data (data with trends) stationary.

     Application: Ideal for data that shows a clear, non-seasonal upward or downward trend. Example: Long-term forecasting of housing prices or population growth

  4. SARIMA Structure: Extends ARIMA by adding seasonal components. It incorporates separate parameters for Trend \((p,d,q)\) and Seasonality \((P,D,Q,s)\), where \(s\) is the periodic cycle length.

     Application: Used for data that repeats patterns at regular intervals during the year. Example: Forecasting electricity demand that peaks during summer months, or quarterly retail sales

  5. SARIMAX Structure: Builds directly on SARIMA by adding external variables, known as exogenous variables(x)

     Application: Highly versatile for scenarios where outside factors drive the variable you are trying to forecast. Example: Forecasting restaurant daily reservations while factoring in marketing spend, weather conditions, or local holidays as exogenous inputs






    

**6.  Load a time series dataset (e.g., AirPassengers), plot the original series,and decompose it into trend, seasonality, and residual components?**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url)

df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

df.columns = ['Passengers']

plt.figure(figsize=(10, 5))
plt.plot(df['Passengers'], label='AirPassengers')
plt.title('Original AirPassengers Time Series')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()

decomposition = seasonal_decompose(
    df['Passengers'],
    model='multiplicative',
    period=12
)

fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

**7.   Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot?**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("NYC_taxi_fare_data.csv")

features = ['fare_amount', 'trip_distance']
X = df[features]

X = X.dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.02,
    random_state=42
)

predictions = iso_forest.fit_predict(X_scaled)

X['Anomaly'] = predictions

n_anomalies = (predictions == -1).sum()
print("Number of anomalies detected:", n_anomalies)

normal = X[X['Anomaly'] == 1]
anomaly = X[X['Anomaly'] == -1]

plt.figure(figsize=(10, 6))

plt.scatter(
    normal['trip_distance'],
    normal['fare_amount'],
    alpha=0.5,
    label='Normal'
)

plt.scatter(
    anomaly['trip_distance'],
    anomaly['fare_amount'],
    marker='x',
    s=100,
    label='Anomaly'
)

plt.xlabel("Trip Distance")
plt.ylabel("Fare Amount")
plt.title("Isolation Forest Anomaly Detection on NYC Taxi Fare Data")
plt.legend()
plt.show()
```

**8.  Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results?**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url)

df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

df.columns = ['Passengers']

model = SARIMAX(
    df['Passengers'],
    order=(1,1,1),
    seasonal_order=(1,1,1,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit()

forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)

forecast_mean = forecast.predicted_mean

conf_int = forecast.conf_int()

future_dates = pd.date_range(
    start=df.index[-1] + pd.DateOffset(months=1),
    periods=forecast_steps,
    freq='MS'
)

plt.figure(figsize=(12,6))

plt.plot(df.index, df['Passengers'],
         label='Historical Data')

plt.plot(future_dates, forecast_mean,
         label='Forecast',
         linestyle='--')

plt.fill_between(
    future_dates,
    conf_int.iloc[:, 0],
    conf_int.iloc[:, 1],
    alpha=0.3,
    label='95% Confidence Interval'
)

plt.title('SARIMA Forecast for AirPassengers Dataset')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()

print("12-Month Forecast:")
print(forecast_mean)
```

**9.   Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib?**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler

data = load_breast_cancer()

X = pd.DataFrame(
    data.data[:, :2],
    columns=data.feature_names[:2]
)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

lof = LocalOutlierFactor(
    n_neighbors=20,
    contamination=0.05
)

y_pred = lof.fit_predict(X_scaled)

X['Anomaly'] = y_pred

num_anomalies = (y_pred == -1).sum()
print("Number of anomalies detected:", num_anomalies)

normal = X[X['Anomaly'] == 1]
anomalies = X[X['Anomaly'] == -1]

plt.figure(figsize=(10, 6))

plt.scatter(
    normal.iloc[:, 0],
    normal.iloc[:, 1],
    alpha=0.6,
    label='Normal Points'
)

plt.scatter(
    anomalies.iloc[:, 0],
    anomalies.iloc[:, 1],
    marker='x',
    s=100,
    label='Anomalies'
)

plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.title('Local Outlier Factor (LOF) Anomaly Detection')
plt.legend()
plt.grid(True)
plt.show()
```

**10.   You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:**

  ● How would you detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN)?

  ● Which time series model would you use for short-term forecasting (ARIMASARIMA / SARIMAX)?

  ● How would you validate and monitor the performance over time?

  ● How would this solution help business decisions or operations?

Here is how I would design and explain the real-time data science workflow to tackle this problem.

### 1. Real-Time Anomaly Detection

For streaming 15-minute interval data, speed and scalability are paramount. Among the choices provided, I would use **Isolation Forest**, adapted for a streaming context using a sliding window approach.

* Isolation Forest because It builds ensembles of random trees and "isolates" observations. Anomalies are isolated closer to the root of the tree, requiring fewer splits. It is computationally efficient, handles high-dimensional data well (important since we have multiple features like region and weather), and scales effectively.
* Why not LOF or DBSCAN?(LOF) relies on nearest-neighbor distance computations, which become computationally expensive and slow as the dataset grows, making it difficult to maintain strict real-time SLA requirements for 15-minute streaming intervals.
* DBSCAN is density-based. Power consumption data often has varying densities depending on the time of day (e.g., tight clusters during off-peak night hours, loose clusters during peak day hours). DBSCAN struggles when cluster densities vary significantly.


* Implementation Strategy: I would engineer "lag features" (e.g., consumption 15 mins ago, 1 hour ago, 1 day ago) and rolling statistics (mean, variance) over a time window. The Isolation Forest would evaluate these rolling features to flag sudden contextual spikes or drops.

### 2. Short-Term Forecasting

I would strictly choose SARIMA (Seasonal Auto-Regressive Integrated Moving Average with eXogenous factors).

* SARIMAX because Power grid demand is highly dependent on two major factors: time and external conditions.
* Seasonality (The "S"): Energy consumption has distinct daily, weekly, and even annual cycles (e.g., people waking up and turning on heaters/ACs, offices closing on weekends). SARIMA components handle these repeating patterns.
* eXogenous Variables (The "X"): This is the deciding factor. Weather conditions (temperature, humidity, cloud cover) heavily dictate energy usage. ARIMA and SARIMA cannot ingest these external features natively. SARIMAX allows us to feed weather and regional data into the model as exogenous variables, drastically improving the accuracy of short-term forecasts.



### 3. Validation and Monitoring

Traditional cross-validation (like k-fold) does not work for time series because it leaks future data into the training set.

* Validation (Walk-Forward Validation) : I would use Time-Series Split (or rolling-origin cross-validation). We train the model on past data (e.g., weeks 1-4) to predict week 5. Then, we expand the training set to include week 5 to predict week 6, moving forward chronologically. This perfectly mimics the production environment.
* Monitoring Metrics: I would track the MAPE (Mean Absolute Percentage Error) and RMSE (Root Mean Squared Error) of the forecasts in real-time.
* Concept Drift Detection: Power consumption behaviors change (e.g., a sudden heatwave, or long-term shifts like increased EV adoption). I would set up monitoring alerts using tools like Evidently AI to track the statistical distribution of the incoming data. If the model's error rate breaches a predefined threshold, or if data drift is detected, it triggers an automated pipeline to retrain the SARIMAX model on the most recent data.

### 4. Business and Operational Impact

Deploying this dual architecture (Anomaly Detection + Forecasting) provides massive, quantifiable value to the power grid operator:

* Grid Stability and Blackout Prevention: By accurately forecasting peak loads via SARIMAX, the operations team knows exactly when to spin up auxiliary power plants (peaker plants) or buy energy from neighboring grids, preventing brownouts.
* Cost Optimization: Storing energy or keeping power plants on standby is incredibly expensive. Accurate short-term forecasting prevents over-generation, saving millions of dollars in wasted operational costs.
* Predictive Maintenance: The Isolation Forest anomaly detector isn't just looking at consumer usage; if a specific region suddenly drops to zero consumption while the forecast predicted high demand, the model immediately alerts operators to a physical fault (like a blown transformer or a downed line), allowing repair crews to be dispatched instantly.
* Dynamic Pricing: With accurate demand curves, the business can implement "Time-of-Use" pricing strategies, incentivizing consumers to shift heavy power usage (like running washing machines or charging EVs) to off-peak hours, flattening the demand curve.
